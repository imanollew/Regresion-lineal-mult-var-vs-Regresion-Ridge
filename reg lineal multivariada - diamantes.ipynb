```python
import pandas as pd
import sklearn
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge
```

```python
#Le damos un formato legible al dataset para que lo interprete pandas
df = pd.read_csv('diamonds.csv', index_col=0) #index_col=0 esta para remover el indice que viene por defecto en el CSV
df.head()
```

```python
#En el dataset tenemos datos que no son numericas, por lo que debemos darles un valor numerico
from sklearn.preprocessing import LabelEncoder
categorical_features = ['cut', 'color', 'clarity']
longitud=len(categorical_features)
le = LabelEncoder()
# convertimos las variables en numeros
for i in range(longitud):
    new = le.fit_transform(df[categorical_features[i]])
    df[categorical_features[i]] = new
df.head()

```

```python
df.describe()
```

```python
#Decimos quien va a ser el output (y) y quienes el input (x)
X = df[['carat', 'depth', 'table', 'x', 'y', 'z', 'clarity', 'cut', 'color']]
Y = df["price"]
print(X)
```

```python
#Separamos el dataset en un 70% para generar un modelo y otro 30% para entrenarlo
#mostramos cuantas filas pertenecen a cada uno
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.30, random_state=40)
print(X_train.shape); print(X_test.shape)
```

```python
pearson_corr = df.corr(method='pearson')
pearson_corr
```

```python
#Utilizamos el modelo de regresion lineal
lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
#Mostramos los coeficientes que permiten predecir
b = lr.coef_
print(b)


print("\nPrecision del modelo R2: ")
print(lr.score(X_train, y_train))



```

```python
# Creamos un dataframe que compare el valor actual con el valor predecido para comparar los valores
close_predictions_lr = lr.predict(X_test)
df = pd.DataFrame({'Actual': y_test, 'Predicted_lr': close_predictions_lr})
df1 = df.tail(25)

print(df1)
```

```python
#Utilizamos el modelo de regresion Ridge
rr = Ridge(alpha=1)
rr.fit(X_train, y_train)
w = rr.coef_
print("Coeficientes W")
print(w)


print("\nPrecision del modelo R2: ")
print(rr.score(X_train, y_train))



```

```python
# Creamos un dataframe que compare el valor actual con el valor predecido para comparar los valores
close_predictions_rr = rr.predict(X_test)
df = pd.DataFrame({'Actual': y_test, 'Predicted_rr': close_predictions_rr})
df2 = df.tail(25)

print(df2)
```
